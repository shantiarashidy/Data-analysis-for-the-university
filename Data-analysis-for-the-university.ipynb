```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sls
```

```python
file_path = r"D:\uni and work\Computer Engineering Scientific Association/categorized_computer_engineering_courses.csv"

df = pd.read_csv(file_path)

df.head()
```

```python
category_counts = df.groupby("Category")["Participants"].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 6))
sls.barplot(x=category_counts.values, y=category_counts.index, palette="viridis")
plt.xlabel("Total Participants")
plt.ylabel("Skill Category")
plt.title("Total Participants by Course Category")
plt.tight_layout()
plt.show()
```

```python
instructor_counts = df.groupby("Instructor")["Participants"].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sls.barplot(x=instructor_counts.values, y=instructor_counts.index, palette="rocket")
plt.xlabel("Total Participants")
plt.ylabel("Instructor")
plt.title("Total Participants per Instructor")
plt.tight_layout()
plt.show()
```

```python
top_courses = df.sort_values(by="Participants", ascending=False)

plt.figure(figsize=(10, 6))
sls.barplot(x=top_courses["Participants"], y=top_courses["Course"], palette="mako")
plt.xlabel("Participants")
plt.ylabel("Course")
plt.title("Courses by Popularity")
plt.tight_layout()
plt.show()
```

```python
df["Category"].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6, 6), startangle=90, cmap="Set3")
plt.title("Distribution of Course Categories")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sls.countplot(y="Instructor", data=df, order=df["Instructor"].value_counts().index, palette="cubehelix")
plt.title("Number of Courses per Instructor")
plt.tight_layout()
plt.show()
```
